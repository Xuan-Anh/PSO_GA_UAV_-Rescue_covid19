```python
import math
ox = []
oy = []
oz = []
for x in range(200):
    ox.append(x)
    y = ((math.e)**(x/30))*1/4
    # y =1/ (1-x/200)
    oy.append(y)
    oz.append(x*0.8*y)
import matplotlib.pyplot as plt
plt.figure(figsize=(5,5))
plt.plot(ox, oy)
plt.show()
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.scatter3D(ox, oy, oz, c=oz,cmap='Reds');
ax.view_init(10, 5)
ax.figure.figsize = (10,10)
plt.show()
```

```python
import numpy as np
def get_travel_time(w):
    g = 9.8
    Wd = 50
    P = 2600
    v0 = 200
    # print(g*(Wd + w)/P)
    ts = np.sin(np.arccos(g*(Wd + w)/P))
    # print(ts)
    ms = np.sin(np.arccos(g*w/P))
    # print(ms)
    travel_time  = ms/ts*v0 - 200
    return travel_time

```

```python
import math
ox = []
oy = []
oz = []
oy_1 = []
for x in range(200):
    ox.append(x)
    y =1/ (1-x/200)
    y_1 = get_travel_time(x)
    oy.append(y)
    oy_1.append(y_1)
    oz.append(x*0.8*y)
import matplotlib.pyplot as plt
plt.figure(figsize=(5,5))
plt.plot(ox, oy, 'k--')
plt.plot(ox, oy_1)


plt.show()
```
