```python
import numpy as np
def get_travel_time(w, Wd,  v0, P):
    g = 9.8
    Wd = 45
    P = 2400
    v0 = 150
    ts = np.sin(np.arccos(g*(Wd + w)/P))
    # print(ts)
    ms = np.sin(np.arccos(g*w/P))
    # print(ms)
    travel_time  = ms/ts*v0 - v0
    return travel_time

```
